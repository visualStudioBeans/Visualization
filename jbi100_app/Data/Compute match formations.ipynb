```python
#data mooi krijgen + Y coordinaten achter elkaar plakken
#data is verrassend simpel, Y is iets in [1,11]
#keeper heb ik weggelaten, als het goed is is dat elke keer speler 1
df_match = df[['id', 'date', 'home_team_goal','away_team_goal','home_player_X1','home_player_X2','home_player_X3','home_player_X4','home_player_X5','home_player_X6','home_player_X7','home_player_X8','home_player_X9','home_player_X10','home_player_X11','away_player_X1','away_player_X2','away_player_X3','away_player_X4','away_player_X5','away_player_X6','away_player_X7','away_player_X8','away_player_X9','away_player_X10','away_player_X11','home_player_Y1','home_player_Y2','home_player_Y3','home_player_Y4','home_player_Y5','home_player_Y6','home_player_Y7','home_player_Y8','home_player_Y9','home_player_Y10','home_player_Y11','away_player_Y1','away_player_Y2','away_player_Y3','away_player_Y4','away_player_Y5','away_player_Y6','away_player_Y7','away_player_Y8','away_player_Y9','away_player_Y10','away_player_Y11']].dropna().copy().reset_index()
homeYs = ['home_player_Y2','home_player_Y3','home_player_Y4','home_player_Y5','home_player_Y6','home_player_Y7','home_player_Y8','home_player_Y9','home_player_Y10','home_player_Y11']
awayYs = ['away_player_Y2','away_player_Y3','away_player_Y4','away_player_Y5','away_player_Y6','away_player_Y7','away_player_Y8','away_player_Y9','away_player_Y10','away_player_Y11']
df_match['home_players_Y'] = df_match[homeYs].astype(int).astype(str).apply(lambda x: '-'.join(x), axis=1)
df_match['away_players_Y'] = df_match[awayYs].astype(int).astype(str).apply(lambda x: '-'.join(x), axis=1)
df_match[['home_players_Y', 'away_players_Y']][1500:1530]
```

```python
#telt achtereenvolgende gelijke getallen
#thanks chatgpt :)
def count_unique_values(concatenated_string):
    values = list(map(int, concatenated_string.split('-')))
    unique_counts = []
    seen = set()

    for value in values:
        if value not in seen:
            unique_counts.append(values.count(value))
            seen.add(value)

    return '-'.join(map(str, unique_counts))
df_match['home_formation'] = df_match['home_players_Y'].apply(count_unique_values)
df_match['away_formation'] = df_match['away_players_Y'].apply(count_unique_values)
df_match.head()
```
